```python
data = [1,2,3,4,5,6,5,1,1,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data, color='#700EBB')     # Represents the frequencies of the data points
plt.xlabel('Representing unique values')
plt.ylabel('Representing frequencies')
plt.title('Histogram plot Example')
plt.grid()

plt.savefig("histogram_test.png")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c='r', alpha=.8)
plt.xlabel('my x axis')
plt.ylabel('my y axis')
plt.title('3D Graph')
plt.show()

plt.savefig('3D Graph.png')
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')

p = figure(title='test flower')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
output_file('test.html')

p = figure(title='test flower')
p.xaxis.axis_label = 'petal_length'
p.yaxis.axis_label = 'petal_width'
p.line(flowers['petal_length'], flowers['petal_width'])
show(p)
```
