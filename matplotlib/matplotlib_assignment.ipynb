## Question 2

What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Scatter plots: A plot that displays the relationship between two variables, where each point represents an observation.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize= (6,4))
plt.scatter(x=x,y=y,c='r', marker='x')
plt.title("Scatter Plot of X and Y Variable")
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.show()
```

## Question 3:

Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots on a single figure, allowing you to compare and visualize multiple datasets or relationships between variables at the same time. The subplot() function takes three arguments: the number of rows of subplots, the number of columns of subplots, and the index of the subplot you want to create.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Create a figure with four subplots arranged in a 2x2 grid.
fig = plt.figure(figsize=(8,7))
ax = fig.subplots(nrows=2, ncols=2)

#Plot the data on each subplot
ax[0,0].plot(x1,y1)
ax[0,0].set_title('The First Line Plot')
#ax[0,0].set_xlabel('X_Label')
ax[0,0].set_ylabel('Y_Label')

ax[0,1].plot(x2,y2)
ax[0,1].set_title('The Second Line Plot')
#ax[0,1].set_xlabel('X_Label')
ax[0,1].set_ylabel('Y_Label')

ax[1,0].plot(x3,y3)
ax[1,0].set_title('The Third Line Plot')
#ax[1,0].set_xlabel('X_Label')
ax[1,0].set_ylabel('Y_Label')

ax[1,1].plot(x4,y4)
ax[1,1].set_title('The Fourth Line Plot')
#ax[1,1].set_xlabel('X_Label')
ax[1,1].set_ylabel('Y_Label')

# Thr Main title for the entire plot
fig.suptitle('The Four Line Plot')

plt.show()
```

## Question 4:

What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


A bar plot is a type of plot that represents categorical data using rectangular bars. 

Bar plots are commonly used to compare and visualize the frequency, count, or proportion of different categories or groups in a dataset.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
#Define fig size
fig = plt.figure(figsize=(8,6))
#Plot figure
plt.bar(company, profit, align='center', color='#33BEFF')
plt.xlabel('Company Name')
plt.ylabel('Profit Level')
plt.title('Top Tech Companies Profit Ranking.')
plt.show()
```

## Question 5: 

What is a box plot? Why is it used? Using the following data plot a box plot.


A box plot, also known as a box and whisker plot, is a statistical representation of the distribution of a dataset. It displays the median, quartiles, and outliers of the data in a way that makes it easy to compare different datasets or identify unusual observations.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# define the figure and subplot object
fig = plt.figure(figsize=(12,6))
ax = fig.subplots(1,2)
#Create the boxplot
ax[0].boxplot(box1)
ax[0].set_title("Boxplot1")
ax[0].set_ylabel('Value')

ax[1].boxplot(box2)
ax[1].set_title("Boxplot2")
ax[1].set_ylabel('Value')

# Thr Main title for the entire plot
fig.suptitle('Box Plots')

plt.show()
```
